```python
%reload_ext autoreload
%autoreload 2
from chromatopy.core import ChromHandler
from chromatopy.readers import ChemstationReader
```

```python
data_path = "/Users/max/Documents/training_course/hao"
detector = "flame ionization detector"

# Read in all data
analyzer = ChromHandler.read(
    path=data_path,
    reader=ChemstationReader,
)

# Plot all data
analyzer.visualize_peaks(detector=detector)
```

## Define analytes of interest

```python
analyzer.add_analyte(
    name="starting material",
    retention_time=6.95,
    detector=detector,
    calibration_factor=0.7578,
)

analyzer.add_analyte(
    name="product 1",
    retention_time=9.53,
    detector=detector,
    calibration_factor=0.992,
)

analyzer.add_analyte(
    name="product 2",
    retention_time=10.01,
    detector=detector,
    calibration_factor=1.0093,
)
```

## Define internal standard

```python
analyzer.add_internal_standard(
    name="internal standard",
    retention_time=7.78,
    detector=detector,
    molecular_weight=246.18,
)
```

## Calculate concentrations

```python
%%capture
analyzer.calculate_concentrations()
```

## Visualize concentrations

```python
# Visualize concentrations
analyzer.visualize_concentrations()
```

## Save data to csv

```python
analyzer.concentrations_to_csv("concentrations.csv")
```
